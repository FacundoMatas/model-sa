```python
import re

def cleanhtml(raw_html):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', raw_html)
    return cleantext
```

```python
print(cleanhtml('<a href="http://twitter.com/download/android" rel="nofollow">Twitter for Android</a>'))
```

```python
def cleanhtmlRow(row):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', row['source'])
    return cleantext
```
