```python
from os import scandir, getcwd
import shutil
import pandas as pd
import json
```

```python
files_path=getcwd()+'\\CSV\\'
file = 'tweets.csv'
print(files_path+file)
```

```python
df = pd.read_csv(files_path+file, sep='\t')
df.describe()
```

```python
# df = df.set_index('id')
# df.head()
```

```python
df=df.sort_values(by=['id'])
df
```

```python
df.drop_duplicates(subset=['id'])
```

```python
#bajarian a 2681 filas
df = df.drop_duplicates(subset=['id'])
```

```python
#Elimino las filas mal cargadas
df=df[df.id.str.contains('[0-9]') == True]
df=df[df.id != '0']
df
```

Elimino tweets en portugues

```python
df=df[df.lang != 'pt']
df
```

```python
#Eliminamos retweets TODOS LOS TWEETS QUE SEAN RETWEETS
# df=df[df.text.str.contains('RT @') == False]
# df


#PARA ELIMINAR SOLAMENTE LA ETIQUETA "RT @" PERO EL MENSAJE SEA UNO MAS:}
df["text"] = df["text"].str.replace("RT @.*:", "")

```
